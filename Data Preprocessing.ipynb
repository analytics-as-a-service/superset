```python
import datetime
```

```python
l=[]
prev_delta=int(data['TransactionDT'][0])
curr_date=datetime.datetime(2021, 3, 10, 15, 10, 30)
l.append(str(curr_date))
for i in range(1,len(data['TransactionDT'])):
    diff = int(data['TransactionDT'][i])-prev_delta
    prev_delta = int(data['TransactionDT'][i])
    temp = curr_date+datetime.timedelta(seconds=diff)
    curr_date = temp
    l.append(str(temp))
```

```python
len(l)
```

```python
data['TransactionTime'] = l
```
